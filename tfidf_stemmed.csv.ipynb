```python
import pandas as pd

# CSV dosyalarını oku
df_investigations = pd.read_csv("islenmis_investigations.csv")
df_car_models = pd.read_csv("islenmis_car_models.csv")

# İlk birkaç satıra göz at
df_investigations.head()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# Eksik değerleri boş metinle doldur
df_investigations['Summary'] = df_investigations['Summary'].fillna("")

# TF-IDF vektörleştirici oluştur
vectorizer = TfidfVectorizer(stop_words='english', max_features=1000)

# TF-IDF matrisini oluştur
tfidf_matrix = vectorizer.fit_transform(df_investigations['Summary'])

# Matriksi DataFrame'e çevir
tfidf_df = pd.DataFrame(tfidf_matrix.toarray(), columns=vectorizer.get_feature_names_out())

# Sonucu incele
tfidf_df.head()
```
